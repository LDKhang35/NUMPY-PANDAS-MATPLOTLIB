### BAI TAP TET
CHO MA TRAN =  \ [
    [99,99,99], \
    [99,99,99], \
    [99,99,99] \
] \
giả sử 0 = O và 1 là x \
nhận đầu vào từ phía X và phía O luân phiên \
cho môt cặp chỉ số \
nếu phía X nhập ((0,0)) thì ma trận trở thành \
[X,99,99], \
[99,99,99], \
[99,99,99] \
nếu phía O nhập ((0,0)) thì yêu cầu nhập lại và nếu không thì điền vào ma trận \ 
thử thách: nếu AI có 3 ô liên tiếp thì dừng trò chơi


```python
import numpy as np

def check_winner(matrix, player):
  """Kiểm tra người chơi đã thắng chưa."""
  for i in range(3):
    if np.all(matrix[i, :] == player) or np.all(matrix[:, i] == player):
      return True
  if np.all(np.diag(matrix) == player) or np.all(np.diag(np.fliplr(matrix)) == player):
    return True
  return False

def play_game():
  """Chơi game Tic-Tac-Toe."""
  matrix = np.full((3, 3), 99)
  current_player = 0

  while True:
    print(matrix)
    row, col = map(int, input(f"Người chơi {current_player} nhập tọa độ (hàng cột): ").split())
    matrix[row, col] = current_player

    if check_winner(matrix, current_player):
      print(f"Người chơi {current_player} đã thắng!")
      break

    if np.all(matrix != 99):
      print("Hòa!")
      break

    current_player = 1 - current_player

if __name__ == "__main__":
    play_game()
```
