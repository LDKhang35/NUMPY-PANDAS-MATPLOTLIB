```python
# 4
import pandas as pd

# Loại bỏ các dòng có giá trị thiếu
df_corr = df[['Campus', 'Attendance (%)', 'Score']].dropna()

# Tính hệ số tương quan Pearson cho từng Campus
correlations_by_campus = df_corr.groupby('Campus').apply(
    lambda group: group['Attendance (%)'].corr(group['Score'])
)

print("Tương quan Attendance vs Score theo Campus:")
print(correlations_by_campus)
```

### Chiến lược
Nếu correlation > 0.6 → Điểm danh ảnh hưởng mạnh đến điểm số → Nên siết chặt điểm danh: chấm điểm danh nghiêm ngặt, phạt vắng mặt.

Nếu correlation thấp (gần 0) → Điểm danh không ảnh hưởng nhiều → Khuyến khích học linh hoạt: học online, gửi bài tập thay vì buộc điểm danh.

Nếu correlation âm (rất hiếm) → Cần xem lại chất lượng lớp học (đi học nhiều mà điểm lại thấp).

```python
# 5
import seaborn as sns
import matplotlib.pyplot as plt

# Vẽ boxplot Score theo Course và Gender
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Course', y='Score', hue='Gender')
plt.title("Boxplot: Score theo Course và Gender")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Tìm outliers dựa trên IQR
outliers = []

# Nhóm theo Course và Gender
grouped = df.dropna(subset=['Score']).groupby(['Course', 'Gender'])

for name, group in grouped:
    Q1 = group['Score'].quantile(0.25)
    Q3 = group['Score'].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outlier_rows = group[(group['Score'] < lower_bound) | (group['Score'] > upper_bound)]
    outliers.append(outlier_rows)

# Gộp các outlier lại thành một DataFrame
outliers_df = pd.concat(outliers)
print("Số lượng outlier:", len(outliers_df))
print(outliers_df[['Student_ID', 'Course', 'Gender', 'Score']].head())
```

```python
# 11
import pandas as pd
import numpy as np

# Đọc file dữ liệu
df = pd.read_csv("customer_data.csv")

# Lọc các dòng đủ điều kiện để huấn luyện
train_data = df.dropna(subset=["Score", "Attendance (%)", "Study_Hours"])

if len(train_data) > 0:
    # Chuẩn bị dữ liệu cho hồi quy
    X = train_data[["Attendance (%)", "Study_Hours"]].values
    y = train_data["Score"].values

    # Thêm cột 1 cho bias
    X_b = np.c_[np.ones((X.shape[0], 1)), X]

    # Tính hệ số beta
    beta = np.linalg.inv(X_b.T @ X_b) @ X_b.T @ y
    print(f"Hệ số hồi quy: y = {beta[0]:.2f} + {beta[1]:.2f}*Attendance + {beta[2]:.2f}*Study_Hours")

    # Dự đoán Score cho các dòng thiếu
    missing = df["Score"].isna() & df["Attendance (%)"].notna() & df["Study_Hours"].notna()
    X_missing = df.loc[missing, ["Attendance (%)", "Study_Hours"]].values
    X_missing_b = np.c_[np.ones((X_missing.shape[0], 1)), X_missing]
    predicted_scores = X_missing_b @ beta
    df.loc[missing, "Score"] = predicted_scores
else:
    print("Không đủ dữ liệu để huấn luyện mô hình hồi quy.")
```

```python
# 12
import pandas as pd
from scipy.stats import skew

# Đọc file CSV
def calculate_skewness(file_path):
    df = pd.read_csv(file_path)
    
    # Loại bỏ giá trị NaN trong cột "Score"
    score_values = df["Score"].dropna()
    
    # Tính skewness
    skewness_value = skew(score_values)
    
    return skewness_value

# Đường dẫn file
dataset_path = "customer_data.csv"
skewness_result = calculate_skewness(dataset_path)

print(f"Skewness: {skewness_result}")
```

```python
# 13
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Đọc file CSV
def plot_pairplot(file_path):
    df = pd.read_csv(file_path)
    
    # Chọn các cột cần vẽ
    selected_columns = ["Score", "Attendance (%)", "Study_Hours"]
    
    # Kiểm tra và chuyển đổi dữ liệu nếu cần
    df_selected = df[selected_columns].dropna()
    df_selected = df_selected.apply(pd.to_numeric, errors='coerce').dropna()
    
    # Vẽ pairplot
    sns.pairplot(df_selected, diag_kind="kde")
    plt.suptitle("Pairplot giữa Score, Attendance (%) và Study_Hours", y=1.02)
    plt.show()

# Đường dẫn file
dataset_path = "customer_data.csv"
plot_pairplot(dataset_path)
```

### Đề xuất mô hình 
mô hình đánh giá kết quả học tập theo hệ thống điểm tổng hợp (Composite Scoring System), trong đó kết hợp nhiều yếu tố để phản ánh năng lực thực tế của học sinh


```python
# 14
import pandas as pd
import matplotlib.pyplot as plt

# Đọc file CSV
def analyze_campus_performance(file_path):
    df = pd.read_csv(file_path)
    
    # Loại bỏ giá trị NaN trong cột "Score" và "Campus"
    df = df.dropna(subset=["Score", "Campus"])
    
    # Chuyển đổi "Score" sang kiểu số
    df["Score"] = pd.to_numeric(df["Score"], errors='coerce')
    
    # Nhóm theo Campus và tính tỷ lệ sinh viên có Score trên 80
    campus_performance = df.groupby("Campus")["Score"].apply(lambda x: (x > 80).mean() * 100)
    
    # Vẽ biểu đồ pie chart
    plt.figure(figsize=(8, 8))
    campus_performance.plot(kind='pie', autopct='%1.1f%%', startangle=90, cmap='Set3')
    plt.ylabel('')
    plt.title("Tỷ lệ sinh viên có Score trên 80 theo Campus")
    plt.show()
    
    return campus_performance

# Đường dẫn file
dataset_path = "customer_data.csv"
campus_performance_result = analyze_campus_performance(dataset_path)

print(campus_performance_result)
```

Thưởng theo tiến bộ cá nhân: Khen thưởng sinh viên có sự cải thiện đáng kể thay vì chỉ dựa vào điểm cao.

```python
# 15
import numpy as np
df = pd.read_csv("customer_data.csv")

# Hàm xác định sinh viên có Score ngoài 2 độ lệch chuẩn
def identify_outliers(df):
    # Tính giá trị trung bình và độ lệch chuẩn của cột "Score"
    mean_score = df["Score"].mean()
    std_score = df["Score"].std()

    # Tìm sinh viên có điểm nằm ngoài 2 độ lệch chuẩn
    lower_limit = mean_score - 2 * std_score
    upper_limit = mean_score + 2 * std_score

    # Lọc ra sinh viên có điểm ngoài giới hạn này
    outliers = df[(df["Score"] < lower_limit) | (df["Score"] > upper_limit)]

    return outliers

# Sử dụng hàm với dataset
outliers = identify_outliers(df)
print(outliers)
```

```python
# 17
import pandas as pd
import numpy as np
from sklearn.impute import KNNImputer
from sklearn.preprocessing import OneHotEncoder, MinMaxScaler

# Đọc dữ liệu
data = pd.read_csv("customer_data.csv")

# Điền giá trị thiếu bằng KNN Imputer
imputer = KNNImputer(n_neighbors=5)
data[['Attendance (%)', 'Study_Hours']] = imputer.fit_transform(data[['Attendance (%)', 'Study_Hours']])

# Điền giá trị thiếu bằng trung vị
data_median = data.copy()
data_median[['Attendance (%)', 'Study_Hours']] = data[['Attendance (%)', 'Study_Hours']].fillna(data[['Attendance (%)', 'Study_Hours']].median())

# So sánh kết quả
print("Dữ liệu sau khi điền bằng KNN:")
print(data[['Attendance (%)', 'Study_Hours']].describe())
print("\nDữ liệu sau khi điền bằng trung vị:")
print(data_median[['Attendance (%)', 'Study_Hours']].describe())

# Mã hóa One-Hot Encoding cho Course
ohe = OneHotEncoder()
courses_encoded = ohe.fit_transform(data[['Course']]).toarray()
course_columns = ohe.get_feature_names_out(['Course'])
data_encoded = pd.DataFrame(courses_encoded, columns=course_columns)

# Chuẩn bị tập dữ liệu đầu vào
features = pd.concat([data_encoded, data[['Attendance (%)', 'Study_Hours']]], axis=1)

# Chuẩn hóa dữ liệu về [0, 1]
scaler = MinMaxScaler()
features = pd.DataFrame(scaler.fit_transform(features), columns=features.columns)

print("Dữ liệu sau khi mã hóa và chuẩn hóa:")
print(features.head())
```

```python
# 18
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Tạo lớp Dataset
class StudentDataset(Dataset):
    def __init__(self, data):
        self.X = data[['Study_Hours', 'Attendance (%)', 'Gender', 'Course', 'Campus']].values
        self.y = data['Score'].values
        
        # Chuyển đổi biến phân loại thành biến số
        self.X[:, 2] = np.where(self.X[:, 2] == 'Male', 1, 0)  # Giới tính
        self.X[:, 3] = pd.factorize(self.X[:, 3])[0]  # Khóa học
        self.X[:, 4] = pd.factorize(self.X[:, 4])[0]  # Campus

        # Chuẩn hóa dữ liệu
        self.scaler = StandardScaler()
        self.X = self.scaler.fit_transform(self.X)

    def __len__(self):
        return len(self.y)

    def __getitem__(self, idx):
        return torch.tensor(self.X[idx], dtype=torch.float32), torch.tensor(self.y[idx], dtype=torch.float32)

# Định nghĩa mô hình ANN
class ANN(nn.Module):
    def __init__(self):
        super(ANN, self).__init__()
        self.fc1 = nn.Linear(5, 32)
        self.fc2 = nn.Linear(32, 16)
        self.fc3 = nn.Linear(16, 8)
        self.fc4 = nn.Linear(8, 1)
        self.relu = nn.ReLU()

    def forward(self, x):
        x = self.relu(self.fc1(x))
        x = self.relu(self.fc2(x))
        x = self.relu(self.fc3(x))
        x = self.fc4(x)
        return x

# Đọc dữ liệu
data = pd.read_csv('customer_data.csv').dropna()
train_data, test_data = train_test_split(data, test_size=0.2, random_state=42)

train_dataset = StudentDataset(train_data)
test_dataset = StudentDataset(test_data)

train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=16, shuffle=False)

# Khởi tạo mô hình, loss function và optimizer
model = ANN()
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Huấn luyện mô hình
epochs = 200
patience = 20
best_loss = float('inf')
counter = 0
train_losses = []

for epoch in range(epochs):
    model.train()
    epoch_loss = 0
    for inputs, targets in train_loader:
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs.squeeze(), targets)
        loss.backward()
        optimizer.step()
        epoch_loss += loss.item()
    
    avg_loss = epoch_loss / len(train_loader)
    train_losses.append(avg_loss)
    
    # Early stopping
    if avg_loss < best_loss:
        best_loss = avg_loss
        counter = 0
    else:
        counter += 1
        if counter >= patience:
            print(f'Early stopping at epoch {epoch}')
            break

# Vẽ biểu đồ loss
plt.plot(train_losses)
plt.title('Loss over epochs')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.show()
```

### Đề xuất
Input Layer: 5 nơ-ron.

Hidden Layer 1: 64 nơ-ron, ReLU.

Hidden Layer 2: 32 nơ-ron, ReLU.

Hidden Layer 3: 16 nơ-ron, ReLU.

Output Layer: 1 nơ-ron (Score).

### Lý do
Tăng số lượng nơ-ron: Việc tăng số lượng nơ-ron trong các tầng ẩn có thể giúp mô hình học được các đặc trưng phức tạp hơn từ dữ liệu.

Số tầng ẩn nhiều hơn: Giúp mô hình có khả năng thể hiện các mối quan hệ phi tuyến tính tốt hơn.

Đặc điểm dữ liệu giáo dục: Dữ liệu giáo dục có thể chứa nhiều yếu tố tác động đến điểm số, vì vậy một kiến trúc phức tạp hơn có thể giúp mô hình dự đoán chính xác hơn.

```python
# 19
from sklearn.metrics import mean_squared_error, r2_score

# Đánh giá mô hình trên tập test
model.eval()
predictions = []
true_values = []

with torch.no_grad():
    for inputs, targets in test_loader:
        outputs = model(inputs)
        predictions.extend(outputs.squeeze().numpy())
        true_values.extend(targets.numpy())

# Tính toán MSE và R²
mse = mean_squared_error(true_values, predictions)
r2 = r2_score(true_values, predictions)

print(f'Mean Squared Error (MSE): {mse}')
print(f'R² Score: {r2}')
```
